```python
def split_and_convert_to_json(string):
    elements = string.split('/')
    json_str = json.dumps(elements)
    return json_str
```

```python
def string_to_dict(string):
    pairs = string.split('/')
    dictionary = {}
    for pair in pairs:
        key, value = pair.split(':')
        dictionary[key] = value
    return dictionary
```
