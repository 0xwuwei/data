```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb 
```

```python
df = pd.read_csv('owid-covid-data.csv')
df.head()
```

```python
df_mx = df.corr()
fig_dims = (32, 32)
fig, ax = plt.subplots(figsize=fig_dims)
sb.heatmap(df_mx, annot=True, cmap='Blues', ax=ax)
```
