```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AttributeCounts.csv')
```

```python
df_0999 = pd.read_csv('0-999.csv')
df_1999 = pd.read_csv('1000-1999.csv')
df_2999 = pd.read_csv('2000-2999.csv')
df_3999 = pd.read_csv('3000-3999.csv')
df_4999 = pd.read_csv('4000-4999.csv')
df_5999 = pd.read_csv('5000-5999.csv')
df_6999 = pd.read_csv('6000-6999.csv')
df_7999 = pd.read_csv('7000-7999.csv')
df_8999 = pd.read_csv('8000-8999.csv')
df_9999 = pd.read_csv('9000-9999.csv')
```

```python
df = pd.concat([df_0999, df_1999,df_2999,df_3999,df_4999,df_5999,df_6999,df_7999,df_8999,df_9999], ignore_index=True)
```

```python
df.head()
```

```python
df['count_of_trait_groups'] = np.where(df[' count']==0, 'zero accessory',
                              np.where(df[' count']==1, 'one accessory',
                              np.where(df[' count']==2, 'two accessory',
                              np.where(df[' count']==3, 'three accessory',
                              np.where(df[' count']==4, 'four accessory',
                              np.where(df[' count']==5, 'five accessory',
                              np.where(df[' count']==6, 'six accessory', 
                              np.where(df[' count']==7, 'seven accessory', 'na'))))))))
```

```python
df.head()
```

```python
df['group_counts'] = df.value_counts(' count')
```

```python
df.to_csv('acc_group_counts.csv', index=False)
```
